## Cleaning Data

```python
# make copies of original pieces of data
survey_cleaned = survey.copy()
```

### Issue 1: 

Drop columns we will not use and Rename columns to have shorter and descriptive column names.

```python
survey_cleaned.columns
```

```python
survey_cleaned.drop( 
    columns = [ 
        'Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week',
        'Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR',
        'Have you lost your job due to the coronavirus outbreak?',
        'Annual bonus+stocks one year ago. Only answer if staying in same country',
        'Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country',
        'Timestamp',
        'Yearly bonus + stocks in EUR'], 
    inplace = True)
```

```python
survey_cleaned.rename(
        columns ={ 
                'Age' : 'age', 
                'Gender' : 'gender', 
                'City' : 'city', 
                'Position ' : 'position', 
                'Total years of experience' : 'experience',
                'Years of experience in Germany' : 'germany_experience', 
                'Seniority level' : 'work_level',
                'Your main technology / programming language' : 'main_tech',
                'Other technologies/programming languages you use often' : 'other_tech',
                'Yearly brutto salary (without bonus and stocks) in EUR' : 'yearly_salary',
                'Number of vacation days' : 'vacation_days', 
                'Employment status' : 'emp_status', 
                'Сontract duration' : 'contract_duration',
                'Main language at work' : 'main_language', 
                'Company size' : 'company_size', 
                'Company type' : 'company_type'}, 
       inplace = True)
```

```python
# list all columns
survey_cleaned.columns
```

### Issue 2: 

Remove values that don't make statistical sense from columns

```python
#check for unique values in the germany experience column
survey_cleaned.germany_experience.unique()
```

```python
replace_dict = {
    '1,5': 1.5,
    '4,5': 4.5,
    '1,7': 1.7,
    '3,5': 3.5,
    '< 1': 1,
    '4 (in Switzerland), 0 (in Germany)': 4,
    '4 month': 4,
    '2,5': 2.5,
    '0,5': 0.5,
    '3 months': 3,
    '3 (in Poland)': 3,
    '<1': 1,
    '⁰':0,
    '0,3': 0.3,
    '-': 0,
    '6 (not as a data scientist, but as a lab scientist)': 6,
    'less than year': 0.9,
}

for key, val in replace_dict.items():
    survey_cleaned.germany_experience = survey_cleaned.germany_experience.replace(key, val)
survey_cleaned.germany_experience = survey_cleaned.germany_experience.astype(float)
```

```python
# clean the main_language column
replace_dict = {
    'English and German' : 'English',
    '50/50' : 'English',
    'Русский' : 'Russian',
    'both' : 'English'
}
for key, val in replace_dict.items():
    survey_cleaned.main_language = survey_cleaned.main_language.replace(key, val)
```

```python
# clean the contract_duration column
replace_dict = {
    '0' : 'Temporary contract'
}
for key, val in replace_dict.items():
    survey_cleaned.contract_duration = survey_cleaned.contract_duration.replace(key, val)
```

```python
# clean the emp_state column
replace_dict = {
'Full-time employee':'full-time',
'Self-employed (freelancer)':'freelancer',
'Company Director':'full-time',
'Founder':'full-time',
'Part-time employee':'part-time',
'working student':'part-time',
'Working Student':'part-time',
"Full-time position, part-time position, & self-employed (freelancing, tutoring)":'full-time',
"full-time, but 32 hours per week (it was my request, I'm a student)":'part-time',
'Werkstudent':'part-time'
}
for key, val in replace_dict.items():
    survey_cleaned.emp_status = survey_cleaned.emp_status.replace(key, val)
```

```python
# clean the experience column
replace_dict = {
    '1,5': 1.5,
    '1 (as QA Engineer) / 11 in total': 11,
    '2,5': 2.5,
    '15, thereof 8 as CTO': 15,
    '6 (not as a data scientist, but as a lab scientist)': 6,
    'less than year': 0.9,
    '383':'38'
}

for key, val in replace_dict.items():
    survey_cleaned.experience = survey_cleaned.experience.replace(key, val)
survey_cleaned.experience = survey_cleaned.experience.astype(float)
```

```python
# work_level Column
replace_dict = {
    'Intern':'intern',
    'Student':'student',
    'no idea, there are no ranges in the firm ':'Senior',
    'No level ' : 'Junior',
    'No level' : 'Junior',
}

for key, val in replace_dict.items():
    survey_cleaned.work_level = survey_cleaned.work_level.replace(key, val)
```

```python
# clean the vacation_days column and convert to float
replace_dict = {
    '(no idea)':'0',
    '30 in contract (but theoretically unlimited)':'30',
    'Unlimited ':'100',
    '~25':'25',
    '23+':'25',
    'Unlimited':'100',
    '24 labour days':'24',
    'unlimited ':100,
    'unlimited':100
}

for key, val in replace_dict.items():
    survey_cleaned.vacation_days = survey_cleaned.vacation_days.replace(key, val)
    # convert vacation number column to float
survey_cleaned['vacation_days'] = survey_cleaned['vacation_days'].astype('float')
```

### Issue 3: 

Fill in null values and drop null rows that are not relevant

```python
survey_cleaned.isnull().sum()
```

```python
# use mean to fill in nulls in age column and change datatype to int

survey_cleaned['age'].fillna(survey_cleaned['age'].mean(), inplace=True)
survey_cleaned['age'] = survey_cleaned['age'].astype('int64')
```

```python
# fill na values in the gender column with a constant value 'Diverse'
survey_cleaned['gender'].fillna('Diverse', inplace=True)
```

```python
# drop rows with null values in position column
survey_cleaned.dropna(subset=['position'], how='all', inplace=True)
```

```python
# use the median to fill in nulls in experience and germany_experience column and change the data type of germany_experience to float
survey_cleaned['experience'].fillna(survey_cleaned['experience'].median(), inplace=True)
survey_cleaned.germany_experience.fillna(survey_cleaned.germany_experience.median,inplace=True)
# survey_cleaned.germany_experience = survey_cleaned.germany_experience.astype('int64')
```

```python
# drop nulls from all the remaining columns
survey_cleaned.dropna(subset=['other_tech'], how='all', inplace=True)
survey_cleaned.dropna(subset=['main_tech'], how='all', inplace=True)
survey_cleaned.dropna(subset=['vacation_days'], how='all', inplace=True)
survey_cleaned.dropna(subset=['contract_duration'], how='all', inplace=True)
survey_cleaned.dropna(subset=['company_type'], how='all', inplace=True)
survey_cleaned.dropna(subset=['contract_duration'], how='all', inplace=True)
survey_cleaned.dropna(subset=['work_level'], how='all', inplace=True)
survey_cleaned.dropna(subset=['emp_status'], how='all', inplace=True)
survey_cleaned.dropna(subset=['main_language'], how='all', inplace=True)
survey_cleaned.dropna(subset=['company_size'], how='all', inplace=True)
```

## Analyzing and Visualizing Data

```python
# Reading the cleaned data to a datframe
final_survey = pd.read_csv('cleaned_it_survey_2020.csv')
```

### Insights:
1. Which positions have many years of experience?

2. Which 5 cities have a high employment rate?

3. Do we have a correlation between salary and toatal years of experience?

4. Do we have a correlation between salary and gender years of experience?

#### Question 1: Which positions have many years of experience?

```python
final_survey.groupby('position')['experience'].mean().sort_values(ascending=False).head(10)
```

```python
# Calculate the mean experience by position and sort in descending order
sorted_means = final_survey.groupby('position')['experience'].mean().sort_values(ascending=False)

# Plot only the top 10 means
sorted_means.head(10).plot(kind='bar', figsize=(12, 8))
plt.title('Top 10 positions by experience', fontsize=20)
plt.ylabel('No. of experience years', fontsize=12)
plt.xlabel('Positions', fontsize=12)
plt.show()
```

*From the insight and the bar chart I can conclude that the top 3 three positions according to the number of experience in years ar; Sales, Tech Leaders and Consultants.*

### Question 2 : Which 5 cities have a high employment rate?

```python
final_survey['city'].value_counts().head(5).plot(kind='pie', figsize=(6,6))
plt.title('Top cities with professional', fontsize=20)
```

### Question 3: Do we have a correlation between salary and total years of experience?

```python
final_survey.plot(x = "experience", y = "yearly_salary", kind ='scatter', figsize=(12,8))
plt.title('Correlation between Salary and Experience', fontsize=16)
plt.ylabel('Salary Amount', fontsize=12)
plt.xlabel('Experience in Years', fontsize=12)
```

*There's a weak positive correlation between salary and experience to mean that people with different years of experience tend to receive the same amount of salary.*

### Question 4: Do we have a correlation between salary and gender years of experience?

```python
# Create boxplot
fig, ax = plt.subplots()
boxprops = dict(linestyle='-', linewidth=2, color='blue')
whiskerprops = dict(linestyle='-', linewidth=2, color='red')
flierprops = dict(marker='o', markersize=8, markerfacecolor='green', alpha=0.5)
medianprops = dict(linestyle='-', linewidth=2, color='orange')
meanprops = dict(marker='s', markersize=8, markerfacecolor='black')
ax.boxplot([final_survey[final_survey['gender'] == 'Male']['yearly_salary'], final_survey[final_survey['gender'] == 'Female']['yearly_salary'], final_survey[final_survey['gender'] == 'Diverse']['yearly_salary']], boxprops=boxprops, whiskerprops=whiskerprops, flierprops=flierprops,
           medianprops=medianprops, meanprops=meanprops)
ax.set_xticklabels(['Male', 'Female', 'Diverse'])
ax.set_ylabel('Salary')

# Show the plot
plt.show()
```
